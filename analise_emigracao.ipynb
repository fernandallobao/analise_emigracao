#Global Emigration Trends (1990-2020)
https://www.kaggle.com/datasets/shreyasur965/global-emigration-trends-1990-2020?select=total-number-of-emigrants.csv


**Este conjunto de dados é ideal para:**

Analisando tendências e padrões globais de migração

Estudar as alterações demográficas nos países de origem

Desenvolvimento de modelos preditivos para fluxos migratórios futuros

Informar políticas de imigração baseadas em evidências

Entendendo o impacto dos eventos históricos na migração

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('total-number-of-emigrants.csv')
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
dataset.loc[(dataset['Entity'] == 'Brazil' )]
```

```python
brazil_data = dataset.loc[(dataset['Entity'] == 'Brazil')] #filta brazil

plt.figure(figsize=(5, 3)) #tamanho do gráfico
plt.plot(brazil_data['Year'], brazil_data['Total number of emigrants']) #fornece os valores para os eixos
plt.xlabel('Year') #nomeia eixo X
plt.ylabel('Total number of emigrants') #nomeia eixo Y
plt.title('Emigration from Brazil (1990-2020)') #título
plt.grid(True) #coloca a grade
plt.show() #imprime
```

```python
brazil_data = dataset.loc[(dataset['Entity'] == 'Brazil')]
plt.figure(figsize=(5, 3))  # tamanho do gráfico
plt.bar(brazil_data['Year'], brazil_data['Total number of emigrants']) #bar plotagem em barras, colunas para o eixo x e y

plt.xlabel('Year')
plt.ylabel('Total number of emigrants')
plt.title('Emigration from Brazil (1990-2020)')
plt.grid(True)  # coloca a grade
plt.xticks(rotation=45)  # rotaciona os anos em 45°
plt.tight_layout() # Adjusts plot parameters for a tight layout.

plt.show()
```

```python
brazil_data = dataset.loc[(dataset['Entity'] == 'Brazil')]

plt.figure(figsize=(5, 3))  # Adjust figure size if needed

# Create the histogram using seaborn
sns.histplot(brazil_data['Total number of emigrants'], kde=False, bins=10)
# kde=False removes the kernel density estimate line
# bins=10 sets the number of bins (you can adjust this)

plt.xlabel('Total number of emigrants')
plt.ylabel('Frequency')
plt.title('Distribution of Emigration from Brazil (1990-2020)')

plt.show()
```

```python
country1 = 'Brazil'
country2 = 'India' # Replace with another country from dataset

country1_data = dataset[dataset['Entity'] == country1]
country2_data = dataset[dataset['Entity'] == country2]

plt.figure(figsize=(10, 6))
plt.plot(country1_data['Year'], country1_data['Total number of emigrants'], label=country1)
plt.plot(country2_data['Year'], country2_data['Total number of emigrants'], label=country2)
plt.xlabel('Year')
plt.ylabel('Total number of emigrants')
plt.title(f'Emigration Comparison: {country1} vs {country2}')
plt.legend()
plt.grid(True)
plt.show()
```
